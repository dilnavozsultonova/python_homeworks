```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.subplot(3, 3, 1)  # Top-left subplot
plt.plot([1, 2, 3], [4, 5, 6])
plt.subplot(3,3,9)
plt.plot([1, 2, 3], [4, 5, 6])
```

```python
x = np.arange(-10,11,1)
y = x ** 2
plt.plot(x,y,color='r',marker='o', linestyle='-.')
plt.show()
```

```python
x = np.arange(-5,6)
y = x ** 2
sizes = np.arange(10,21)
plt.scatter(x,y, s=sizes, c=np.arange(0,11), cmap="viridis")
plt.show()
```

```python
categories = ['Red', 'Blue', 'Green', 'Yellow']
values = [10, 15, 7, 12]

plt.bar(categories, values, color=[c.lower() for c in categories], width=0.5, label=np.array(["1","2","3","4"]))
plt.title('Basic Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
import numpy as np

categories = ['A', 'B', 'C', 'D']
values1 = [10, 15, 7, 12]
values2 = [12, 18, 10, 15]

x = np.arange(len(categories))

w = 0.2
plt.bar(x-w/2, values1, width=w, label='Dataset 1', color='blue')
plt.bar(x+w/2, values2, width=w, label='Dataset 2', color='green')

plt.xticks(x, categories)
plt.title('Grouped Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values1 = [10, 15, 7, 3]
values2 = [5, 10, 3, 8]

plt.bar(categories, np.array(values2) + np.array(values1), label='Dataset 1', color='blue')
plt.bar(categories, values2, label='Dataset 2', color='green')

plt.title('Stacked Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
colors = ['red', 'blue', 'green', 'purple']
widths = [0.4, 0.6, 0.8, 1.0]

plt.bar(categories, values, color=colors, width=widths, edgecolor=['blue','yellow'])
plt.title('Bar Chart with Custom Colors and Widths')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=20, color='purple', alpha=0.9, edgecolor='black')
plt.title('Normalized Histogram')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
np.random.seed(42)
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(1, 1, 1000)

plt.hist(data1, bins=20, alpha=1, label='Dataset 1', color='blue', edgecolor='black')
plt.hist(data2, bins=20, alpha=0.5, label='Dataset 2', color='red', edgecolor='black')

plt.title('Histogram with Multiple Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
data = np.random.uniform(0, 10, 50000)

plt.hist(data, bins=15, color='cyan', edgecolor='black', alpha=0.8, rwidth=1)
plt.title('Customized Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
 
ax.plot3D(x, y, z, c='g')
ax.set_title('3D line plot geeks for geeks')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)

xx, yy = np.meshgrid(x, y)

zz = np.sin(np.sqrt(xx**2 + yy**2))

ax.plot_surface(xx, yy, zz, cmap='Blues')
```
